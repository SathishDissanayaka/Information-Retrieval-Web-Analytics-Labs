```python
#quiz 1
teams = ["India", "England", "NZ", "Aus"]
captains = ["Kohli", "Root", "Williaamson", "Smith"]
# Pair them as tuples
print(list(zip(teams, captains)))
# Create a dictionary from the two lists
print(dict(zip(teams, captains)))

#quiz 2(list comprehension)
books = ["textbooks", "exercise books", "story book", "drawing books"]
prices = [100, 60, 90, 70]
quantities = [3, 2, 1, 4]

for book, price, quantity in zip(books, prices, quantities):
    total = price * quantity
    print(f"You bought {quantity} {book} for ${total}")
    #also can use just print(total)

#quiz 3
list1 = [2, 4, 6, 8, 10]
result = [x + 10 for x in list1]
print(result)

#quiz 4 (with reversing)
list1 = [10, 20, 30, 40]
list2 = ["Apples", "Mangoes", "Oranges", "Grapes"]
for a, b in zip(list1, reversed(list2)):
    print(a, b)

#quiz 5 (Multi dimentional Array Indexing)
list1 = ["a", "b", ["c", ["d", "e", ["f", "g"], "k"], "l"], "m", "n"]
sub_list = ["h", "i", "j"]
list1[2][1][2].extend(sub_list)
print(list1)

#quiz 6(list comprehension)
#expression for name_x in listName(condition)
list1 = [10, 15, 20, 15, 32, 54, 15]
result = [x for x in list1 if x != 15]
print(result)

#quiz 7
#update() function
#dictionary is a key value pair
dict_1 = {'John': 15, 'Rick': 10, 'Misa': 12}
dict_2 = {'Bonnie': 18, 'Rick': 20, 'Matt': 16}
dict_1.update(dict_2)
print(dict_1)

#quiz 8
d = {0: 0, 1: 1, 2: 2, 3: 3}
d[4] = d.pop(0)
print(d)

#quiz 9
country = ["USA", "France", "India"]
capital = ["Washington D.C.", "Paris", "New Delhi"]
result = dict(zip(country, capital))
print(result)

#quiz 10
My_dict = {
    "Fruit": "Pear",
    "Vegetable": "Carrot",
    "Pet": "Cat",
    "Book": "Moby dick",
    "Crystal": "Amethyst"
}
keysToRemove = ["Book", "Crystal"]

for key in keysToRemove: #removing multiple items
    My_dict.pop(key, None)

print(My_dict)

#quiz 11
sub_dict = {'math': 100, 'chem': 98, 'sci': 100, 'eng': 100}
key_to_extract = {'math', 'chem', 'sci'}
result = {k: sub_dict[k] for k in key_to_extract}
print(result)

#quiz 12
list1 = [12, 15, 32, 42, 55, 75, 122, 132, 150, 180, 200]
result = [x for x in list1 if x % 5 == 0]
print(result)

#quiz 13
#python loop
numbers = [12, 75, 150, 180, 145, 525, 50]
for num in numbers:
    if num > 500:
        break #exit the loop
    if num > 150:
        continue # skip an iteration
    if num % 5 == 0:
        print(num)

#quiz 14
#isupper(),isdigit(),islower() functions
text = "@W3Resource.Com"
upper = lower = digit = special = 0

for ch in text:
    if ch.isupper():
        upper += 1
    elif ch.islower():
        lower += 1
    elif ch.isdigit():
        digit += 1
    else:
        special += 1
print("Upper case characters:", upper)
print("Lower case characters:", lower)
print("Number case:", digit)
print("Special case characters:", special)

#quiz 15
n = 5
num = ''
total = 0

for i in range(n):
    num += '2'               # Build the number like 2, 22, 222, ...
    total += int(num)        # Convert it to integer and add to total
    print(num, end='+')      # Print the current term with a +

print(f"\nSum of above series is: {total}")




```
